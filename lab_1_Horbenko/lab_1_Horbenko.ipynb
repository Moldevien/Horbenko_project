## Написання та запуск коду. Налаштування PyCharm-Git-GitHub середовища 

Для ефективної, зокрема командної, роботи потрібно створений проект зв'язати з системою контроля версій (СКВ).  
Існують різні СКВ, найбільш популярною з яких наразі є [Git](https://git-scm.com/).  
Для того, щоби налаштувати середовище PyCharm-Git-GitHub необхідно:

1. Завантажити і налаштувати СКВ Git на своєму локальному комп'ютері.

1. Створити репозиторій на GitHub у своєму акаунті.

Наступний крок -- зв'язування його з віддаленим репозиторієм, який знаходиться на GitHub.

### Зв'язування створеного python-проєкта з відддаленим Git-репозиторієм

Зверніть увагу, що пункт меню змінив назву з VCS на Git.
Далі потрібно скопіювати посилання з GitHub-репозиторію і вставити її у віповідне поле в PyCharm. При цьому слід враховувати протокол, за яким налаштована взаємодія Git-на локальному комп'ютері з віддаленим репозиторієм: SSH або HTTPS.

### Робота з віддаленим репозиторієм

Будь-які зміни в проєкті можуть бути додані в індекс Git-репозиторію і закомічені у репозиторій на локальній машині. Після того копія локального репо може бути записана у віддалений репо на GitHub.
Перелік необхідних команд Git, якщо викоростовувати для цього консоль, може бути таким:

* `git status` - перевірка статусу файлів;

* `git add .` - додавання всіх новостворених файлів у індекс;

* `git commit -m "коментар до коміту"` - додавання змін у локальний Git-репозиторій;

* `git push -u origin master` - відсилання поточної локальної копії Git-репо на GitHub-репозиторій.

Початківці це можуть зробити простіше через меню PyCharm.

Після цього ми будемо мати налаштоване середовище для ефективного написання та відлагодження коду на Python з можливістю збереження всіх версій проєкту як у локальному Git-середовищі, так і у віддаленому на GiHub.

## Відповіді на контрольні запитання

1. Відмінність інтерпретатора від компілятора:

- Інтерпретатор виконує код рядок за рядком, перекладаючи його безпосередньо під час виконання. Наприклад, Python використовує інтерпретатор.
- Компiлятор спочатку перекладає весь вихідний код у машинний код (зазвичай у вигляді виконуваного файлу), після чого програма виконується. 

2. Що таке IDE PyCharm:

PyCharm — це інтегроване середовище розробки (IDE) для Python, розроблене компанією JetBrains. Воно надає інструменти для написання, відладки, тестування та розгортання Python-коду, підтримує інтеграцію з Git, SQL базами даних, має автозаповнення коду та потужні інструменти рефакторингу.

3. Що таке система контролю версій (СКВ) і для чого вона призначена? Відмінність між SVN та GIT:

Система контролю версій (СКВ) — це інструмент для відстеження змін у вихідному коді, що дозволяє зберігати попередні версії файлів, співпрацювати між розробниками та легко керувати версіями проєкту.  
Відмінність між SVN та GIT:
- SVN (Subversion) — централізована система, де всі розробники працюють із однією основною копією репозиторію на сервері.
- Git — розподілена система, де кожен розробник має свою повну копію репозиторію і може працювати локально без постійного підключення до сервера.

4. Що відбувається при виконанні команди `git add .`:

Команда `git add .` додає всі нові, змінені або видалені файли з поточної директорії та всіх її піддиректорій до індексу (стейджинг-ареа), підготовлюючи їх до наступного коміту.

5. Що відбувається при виконанні команди `git commit`:

Команда `git commit` створює знімок (snapshot) змін, які були додані до індексу (стейджинг-ареа), і зберігає цей знімок у локальному репозиторії з повідомленням про зміни (коментарем). Цей знімок можна пізніше надіслати до віддаленого репозиторію.

6. Команда для розміщення локального Git-репозиторію на GitHub:

git push "ім'я репозиторію" "гілка"
